```python

import kagglehub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import os

# STEP 1: Download dataset
path = kagglehub.dataset_download("ziya07/blockchain-enabled-6g-microgrid-dataset")

# STEP 2: Load first CSV or Excel file
df = None
for root, dirs, files in os.walk(path):
    for file in files:
        full_path = os.path.join(root, file)
        if file.endswith(".csv"):
            df = pd.read_csv(full_path)
            break
        elif file.endswith(".xlsx") or file.endswith(".xls"):
            df = pd.read_excel(full_path)
            break

# STEP 3: Check if dataset was loaded
if df is None:
    print("‚ùå No dataset file found!")
else:
    print("‚úÖ Dataset loaded!")
    print(df.head())




# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(15, 12))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with better formatting
sns.heatmap(
    corr,
    cmap=cmap,
    annot=True,
    fmt=".2f",
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.75},
    annot_kws={"size": 8}
)

# Rotate axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(fontsize=9)
plt.title('Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```
